```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Housing_Prices3')
```

```python
data.info()
```

```python
data["Zipcode"]=data['Zipcode'].astype('object')
```

```python
data.info()
```

```python
data['No of Times Visited'].unique()
```

```python
d={'None':0,"Four":4,'Once':1,'Twice':2,"Thrice":3}
```

```python
data['No of Times Visited']=data['No of Times Visited'].map(d)
```

```python
data['No of Times Visited'].head(10)
```

```python
data['ever renovated']=np.where(data['Renovated Year']==0,"No","Yes")
```

```python
data.head()
```

```python
def year(value):
    return value.split()[-1]


data['Sale_Year']=data['Date House was Sold'].map(year)
```

```python
data.head()
```

```python
data.info()
```

```python
data["Sale_Year"]=data['Sale_Year'].astype('int64')
```

```python
data.info()
```

```python
data['Renovation_Years']=np.where(data["ever renovated"]=="Yes",abs(data['Sale_Year']-data['Renovated Year']),0)
```

```python
data.head()
```

```python
data.drop(columns=['Date House was Sold','Renovated Year','Sale_Year'],inplace=True)
```

```python
data.head()
```

```python
data.to_csv("Housing_Prices4")
```
