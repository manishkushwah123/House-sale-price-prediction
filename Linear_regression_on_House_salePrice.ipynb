```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
plt.style.use("dark_background")
warnings.filterwarnings(action='ignore')
```

```python
data=pd.read_csv('transformed_Housing_Prices.csv')
```

# Scaling the dataset

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
Y=data['Sale_Price']
X=scaler.fit_transform(data.drop(columns=['Sale_Price']))
X=pd.DataFrame(data=X,columns=data.drop(columns=['Sale_Price']).columns)
X.head()
```

# Checking and removing multicollinearity

```python
X.corr()
```

```python
k=X.corr()
z=[[str(i),str(j)] for i in k.columns for j in k.columns if(k.loc[i,j]>abs(0.5))&(i!=j)]
z,len(z)
```

# Calculating VIF

```python
#importing Variance_inflation_Factor function from the Statsmodels
from statsmodels.stats.outliers_influence import variance_inflation_factor 
vif_data=X


## calculating VIF values for each Column
VIF=pd.Series([variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF
```

```python
## create a function to remove multicollinearity 
def MC_remover(data):
    vif=pd.Series([variance_inflation_factor(data.values, i) for i in range(data.shape[1])],index=data.columns)
    if vif.max()>5:
        print(vif[vif==vif.max()].index[0],'has been removed')
        data =data.drop(columns=[vif[vif==vif.max()].index[0]])
        return data
    else:
        print('No Multicollinearnity present anymore')
        return data
```

```python
for i in range(7):
    vif_data=MC_remover(vif_data)

vif_data.head()
```

# Calculate VIF for Remaing column

```python

VIF=pd.Series([variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF,len(vif_data.columns)
```

# Train/Test Data

```python
x=vif_data
y=data['Sale_Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=101)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression(normalize=True)
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
predictions=lr.predict(x_test)
lr.score(x_test,y_test)
```

# 1.Residuals

```python
residuals=preditions-y_test
residual_table=pd.DataFrame({'residuals':residuals,'predictions':predictions})
residual_table=residual_table.sort_values(by='predictions')
```

```python
z=[i for i in range(int(residual_table['predictions'].max()))]
k=[0 for i in range(int(residual_table['predictions'].max()))]
```

```python
plt.figure(dpi=130,figsize=(17,7))
plt.scatter(residual_table['predictions'],residual_table['residuals'],color='red',s=2)
plt.plot(z,k,color='green',linewidth=3,label='regressionline')
plt.ylim(-800000,800000)
plt.xlabel('fitted points(ordered by preditions)')
plt.ylabel('residuals')
plt.title('residuals plot')
plt.legend()
plt.show()
```

```python
plt.figure(dpi=130,figsize=(10,7))
plt.hist(residual_table['residuals'],color='red',bins=200)
plt.xlabel('residuals')
plt.ylabel('frequency')
plt.title('residuals distribution plot')
plt.legend()
plt.show()
```

# Model Coefficients

```python
coefficients_table=pd.DataFrame({'column':x_train.columns,
                                'coefficients': lr.coef_})
coefficients_table=coefficients_table.sort_values(by='coefficients')
```

```python
plt.figure(dpi=120,figsize=(8,6))
x=coefficients_table['column']
y=coefficients_table['coefficients']
plt.barh(x,y)
plt.xlabel("coefficients")
plt.ylabel('variables')
plt.title("normalalized coefficient plot")
plt.show()
```
